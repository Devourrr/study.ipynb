- 파이썬 코드를 작성하다보면 반복적으로 사용되는 코드가 나옵니다.
이 경우 코드를 단순히 복사한 후 붙여넣기 하는 것보다 해당 코드에 '이름표'를 붙여두고 이름표로 코드를 대신할 수 있는데 이 개념이 바로 함수입니다.

메모리에 새로 할당됩니다.

- 마지막으로 함수호출의 3단계는 '함수가 리턴'되는 것입니다
- 함수가 리턴되면 코드의 실행 흐름이 다시 함수를 호출했던 위치로 돌아갑니다.
- 동시에 return이라는 키워드가 리턴하는 값을 갖고 돌아갑니다.

해당 위치에 7이 리턴됐습니다.
- 따라서 result라는 변수는 함수에서 리턴된 값인 7을 바인딩합니다.
- 함수가 리턴되는 순간 함수 내에서 선언된 변수인 c,d,ret는 모두 소멸합니다.

- 함수가 리턴되면 함수 내에서 선언된 변수가 모두 없어지므로 함수 바깥쪽에서 해당 변수인 c,d,ret를 print할 수 없습니다

- 함수 기본 인자
- 함수는 함수 호출부로부터 인자(Parameter)를 전달받고 그 값을 사용해서 함수를 실행합니다.
- 함수가 호출되면서 인자로 1,2,3을 전달합니다.
- 함수의 인자 a,b,c에는 각각 1,2,3이 바인딩되고 그 값이 차례로 화면에 출력됩니다

````python
def myprint(a,b,c):
    print(a)
    print(b)
    print(c)
    
myprint(1,2,3)
````

- 함수를 호출할 때 인자의 전달 순서를 변경할 수 있습니다.
- 함수에 정의된 변수와 함께 바인딩 될 값을 직접 지정하는 겁니다.
- c에는 3을, b에는 2를, a에는 1을 전달합니다

````python
myprint(c=3,b=2,a=1)
````

- 함수를 정의할 때 함수의 인자로 기본ㄱ밧을 설정할 수 있는데 이를 기본함수 인자라고 합니다.
- myprint함수에서 c인자에 대해 기본값을 설정해 보겠습니다

````python
def myprint(a,b,c='default'):
    print(a)
    print(b)
    print(c)
    
myprint(1,2,3)
myprint(1,2)
````

- 함수가 정의될 때 파라미터는 3개이지만 함수를 호출할 때는 2개만 넘겨줬습니다
- 이 경우 기본 인자 값이 설정된 c는 "default"라는 문자열을 인자로 사용합니다

# 데이터 정렬
- 데이터 정렬에 대해 알아보겠습니다
- 해결하고자하는 문제는 리스트에 정숫값이 있는데 이를 오름차순으로 정렬하는 문제를 해결하고자 해봅시다.

````python
data = [3,2,1,5,4]

sorted_data = sorted(data)
print(data)
print(sorted_data)
````

- sorted()내장 함수를 호출하면 원본 데이터는 그대로 유지하고 정렬된 데이터 객체를 추가로 리턴합니다
- 원본 데이터, 정렬된 데이터 순서대로 출력합니다.
- 파이썬 리스트는 sort()라는 정렬 메서드를 갖고 있습니다.
- sort()메서드를 사용하면 리스트의 데이터를 직접 정렬해버립니다.
- 원본을 유지해야 하는 경우에는 앞서 배운 sorted() 내장 함수를 사용해야 합니다.

````python
data.sort()
print(data)
````

- 기본 데이터에 대한 정렬을 알아봤습니다.
- 데이터의 구조가 조금 복잡한 경우에 대한 정렬을 생각해봅시다.
- 주식 종목 3개에 대해서 각 종목의 종목 코드와 PBR (Price Book-value Ratio)이 튜플로 구성된 경우를 생각해봅시다
- 여기서 PBR이 낮은 순으로 정렬하고 싶습니다.

````python
data = [("000060", 8.25), ("000020", 5.75), ("039490", 1.3)]
data.sort()
print(data)
````

- 리스트에 sort()메서드를 사용하면 튜플의 첫번째 값을 기준으로 정렬합니다.
- sort() 메서드가 티커를 기준으로 정렬하여 우리의 의도(PBR기준 정렬)대로
정렬되지 않았음을 알 수 있습니다.

````python
data = [("000060", 8.25), ("000020", 5.75), ("039490", 1.3)]

def 정렬규칙(x): # x는 튜플 타입 객체
    return x[1] # 튜플의 1번 값을 리턴

data.sort(key=정렬규칙)
print(data)
````

- 정렬규칙이라는 함수를 정의하고 이를 sort()메서드의 key라는 이름의 인자로 전달해보겠습니다.
- 정렬규칙 함수의 기능은 어떤 입력(x)이 들어오면 그 입력의 1번 값(x[1])을 리턴합니다.
- x인자로 리스트의 데이터인 튜플(각 종목에 대한 정보)이 들어올 것으로 가정하고
튜플의 1번위치에 PBR이 저장되어 있기 때문에 1번 위치의 값을 리턴하도록 한겁니다

- 출력값을 확인해보면 PBR 값을 기준으로 데이터가 정렬된 것을 확인할 수 있습니다.
### 정렬코드는 매우 자주 사용되는 코드이므로 외울 때까지 코딩해보세요

- 정렬규칙이라는 함수가 매우 간단한데 그 함수를 정의해줘야 한다는 점이 귀찮은 부분일 수 있습니다.
- 함수가 너무 간단하다보니 함수에 이름을 붙이지않고 간단히 함수의 역할만 기술하고 싶습니다.
### 이 때 사용하는 것이 '이름없는함수'인 lambda 함수입니다

````python
data = [("000060", 8.25), ("000020", 5.75), ("039490", 1.3)]
data.sort(key=lambda x:x[1])
print(data)
````

- sort()메서드의 key인자로 lambda 함수 객체를 전달합니다.
- lambda 함수는 x를 입력받아 x[1]을 리턴하는 기능을 가집니다.
- 출력값을 확인하면 lmabda함수를 사용해도 동일하게 정렬된 것을 확인할 수 있습니다.
